# 1. 그래프

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sign1_file_path = 'sign1.csv'
data1 = pd.read_csv(sign1_file_path)

x = data1['height']

y1 = data1['distance']
plt.figure(figsize=(10, 5))
plt.plot(x, y1, marker='o')
plt.xlabel('Height')
plt.ylabel('Distance')
plt.title('Sign1 - Height vs Distance')
plt.grid(True)
plt.show()

y2 = data1['x']
plt.figure(figsize=(10, 5))
plt.plot(x, y2, marker='o', color='r')
plt.xlabel('Height')
plt.ylabel('X')
plt.title('Sign1 - Height vs X')
plt.grid(True)
plt.show()

y3 = data1['y']
plt.figure(figsize=(10, 5))
plt.plot(x, y3, marker='o', color='g')
plt.xlabel('Height')
plt.ylabel('Y')
plt.title('Sign1 - Height vs Y')
plt.grid(True)
plt.show()
```

# 2. Best Fitting Curve

```python
# 'height'와 'distance' 데이터 추출
x = data1['height']
y = data1['distance']

# 2차 다항식 적합 (quadratic fit)
coefficients = np.polyfit(x, y, 4)  # 2는 다항식의 차수를 의미합니다.
polynomial = np.poly1d(coefficients)

# 적합 곡선에 대한 y 값 계산
x_fit = np.linspace(x.min(), x.max(), 100)
y_fit = polynomial(x_fit)

# 데이터와 적합 곡선 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'o', label='Original data')
plt.plot(x_fit, y_fit, '-', label='Fitted curve')
plt.xlabel('Height')
plt.ylabel('Distance')
plt.title('Height vs Distance with Best Fitting Curve')
plt.legend()
plt.grid(True)
plt.show()

# 적합 곡선의 계수 출력
print(f"Fitted polynomial coefficients: {coefficients}")
```

```python
coefficients = [2.89171349e-05, -8.26735825e-03, 8.98101509e-01, -4.57842714e+01, 1.07120384e+03]
polynomial = np.poly1d(coefficients)
height = 55
print(polynomial(height))
```
